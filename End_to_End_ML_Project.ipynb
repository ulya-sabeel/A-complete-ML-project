```python
#split the features and target variable

X = data[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']]
y = data['Price']
```

```python
#split into train and test data subsets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Build and Train the Model**

```python
#Random Forests work by training many Decision Trees on random subsets of the features, then averaging out their predictions.
#Ensemble learning
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=100, random_state=42) #100 Decision trees
forest_reg.fit(X_train,y_train)
```

```python
# let's check the predictions on a few training instances
some_data = X_test[:5]
some_labels = y_test[:5]

print("Predictions:", forest_reg.predict(some_data))

print("Labels:", list(some_labels)) #Compare against the actual values #predictions not accurate
```

```python
pred_RF = forest_reg.predict(X_test)
```

```python
import holoviews as hv
hv.extension('bokeh')

import hvplot.pandas
import pandas as pd, numpy as np
pd.DataFrame({'True Values': y_test, 'Predicted Values': pred_RF}).hvplot.scatter(x='True Values', y='Predicted Values')
```

**Evaluate the model using several metrics**

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score

def cross_val(model): ##cross validation
    pred = cross_val_score(model, X, y, cv=10)
    return pred.mean()

def print_evaluate(true, predicted):  ## print the results
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
    print('__________________________________')
    
def evaluate(true, predicted): # evaluation metrics
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return mae, mse, rmse, r2_square
```

```python
test_pred_RF = forest_reg.predict(X_test)
train_pred_RF = forest_reg.predict(X_train)

print('Test set evaluation:\n_____________________________________')
print_evaluate(y_test, test_pred_RF)
print('Train set evaluation:\n_____________________________________')
print_evaluate(y_train, train_pred_RF)

res_df = pd.DataFrame(data=[["Random Forest Regressor", *evaluate(y_test, test_pred_RF) , cross_val(RandomForestRegressor(n_estimators=100, random_state=42))]], 
                          columns=['Model', 'MAE', 'MSE', 'RMSE', 'R2 Square', "Cross Validation"])
```

**Apply hyperparameter optimization**

```python
from sklearn.model_selection import RandomizedSearchCV #Random Search HPO
from scipy.stats import randint

#n_estimators = number of trees in the forest
#max_features = max number of features considered for splitting a node

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=5),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs, n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(X_train,y_train)
```

```python
#The best hyperparameter combination found:
rnd_search.best_params_
```

```python
rnd_search.best_estimator_
```

```python
final_model = rnd_search.best_estimator_
```

```python
# let's check the predictions on a few training instances
some_data = X_test[:5]
some_labels = y_test[:5]
print("Predictions:", final_model.predict(some_data))
print("Labels:", list(some_labels)) #Compare against the actual values #predictions not accurate
```

```python
pred_final = final_model.predict(X_test)
test_pred_final = final_model.predict(X_test)
train_pred_final = final_model.predict(X_train)

print('Test set evaluation:\n_____________________________________')
print_evaluate(y_test, test_pred_final)
print('Train set evaluation:\n_____________________________________')
print_evaluate(y_train, train_pred_final)

res_df_final = pd.DataFrame(data=[["Random Forest Regressor Final", *evaluate(y_test, test_pred_final) , cross_val(RandomForestRegressor(max_features=3, n_estimators=180, random_state=42))]], 
                          columns=['Model', 'MAE', 'MSE', 'RMSE', 'R2 Square', "Cross Validation"])
```
